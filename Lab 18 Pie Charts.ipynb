### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python
##Write your code here
import pandas as pd
import matplotlib.pyplot as plt

# Select the relevant column
df_db = df[['DatabaseWantToWorkWith']].dropna().copy()

# Split multiple databases into separate rows
df_db = df_db.assign(Database=df_db['DatabaseWantToWorkWith'].str.split(';')).explode('Database')
```

```python
top5_db = df_db['Database'].value_counts().head(5)
```

```python
plt.figure(figsize=(8,8))
plt.pie(
    top5_db, 
    labels=top5_db.index, 
    autopct='%1.1f%%',  # show percentage
    startangle=140
)
plt.title("Top 5 Databases Respondents Want to Work With")
plt.show()
```

The `DevType` column lists the developer types for respondents. We’ll examine the distribution by showing the top 5 developer roles in a pie chart.



```python
##Write your code here
import pandas as pd
import matplotlib.pyplot as plt

# Select the relevant column
df_dev = df[['DevType']].dropna().copy()

# Split multiple roles into separate rows
df_dev = df_dev.assign(Role=df_dev['DevType'].str.split(';')).explode('Role')
```

### Task 3: Analyzing and Interpreting Composition

##### 3.1 Pie Chart of `Respondents` Most Admired Programming Languages


The `LanguageAdmired` column lists the programming languages respondents admire most. Create a pie chart to visualize the top 5 admired languages.



```python
##Write your code here
import pandas as pd
import matplotlib.pyplot as plt

def create_pie_chart(df, column_name, top_n=5, title="Pie Chart"):
    # Select and drop missing
    df_col = df[[column_name]].dropna().copy()
    
    # Split multiple answers and explode
    df_col = df_col.assign(Item=df_col[column_name].str.split(';')).explode('Item')
    
    # Count top items
    top_items = df_col['Item'].value_counts().head(top_n)
    
    # Pie chart
    plt.figure(figsize=(8,8))
    colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0']
    plt.pie(
        top_items,
        labels=top_items.index,
        autopct='%1.1f%%',
        startangle=140,
        colors=colors[:top_n],
        explode=(0.1,) + (0,)*(top_n-1)
    )
    plt.title(title)
    plt.show()
```

```python
create_pie_chart(df, 'LanguageAdmired', top_n=5, title='Top 5 Most Admired Languages')
```

##### 3.2 Pie Chart of Tools Used for AI Development


Using the `AIToolCurrently` Using column, create a pie chart to visualize the top 5 tools developers are currently using for AI development.



```python
##Write your code here
create_pie_chart(df, 'AIToolCurrently Using', top_n=5, title='Top 5 tool')
```

##### 3.3 Pie Chart for Preferred Web Frameworks


The `WebframeWantToWorkWith` column includes web frameworks that respondents are interested in working with. Visualize the top 5 frameworks in a pie chart.
